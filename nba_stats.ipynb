## Prédiction des performances

### Arbre de décision 

Peut on prédire si les Lakers atteignent les finales NBA à partir de leur stats en saison régulière ? (pour une saison donnée)

#### Utilisation de sklearn

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
liste_features = ['WIN_PCT', 'CONF_RANK', 'DIV_RANK','FG_PCT', 'FG3_PCT', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'PF', 'STL', 'TOV', 'BLK', 'PTS', 'PTS_RANK']

#Passer de dataframe à array
stats_desc = np.array(team_stats_df[liste_features])

stats_label = np.array(team_stats_df['NBA_FINALS_APPEARANCE'])
stats_label = np.where(pd.isnull(stats_label), 'NO_APPEARANCE', stats_label)



# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(stats_desc, stats_label, test_size=0.2, random_state=42)

# Créer un objet de l'algorithme d'arbre de décision
clf = DecisionTreeClassifier()

# Entraîner l'arbre de décision sur les données d'entraînement
clf.fit(X_train, y_train)

# Prédire les étiquettes des données de test
predictions = clf.predict(X_test)

# Calculer l'exactitude des prédictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

```python
# Afficher l'arbre de décision
plt.figure(figsize=(35, 30))
plot_tree(clf, filled=True, feature_names=np.unique(liste_features), class_names=np.unique(stats_label))
plt.show()
```

```python
# VALIDATION CROISEE

sum_accuracy = 0

for i in range (10) : 
    # Diviser les données en ensembles d'entraînement et de test
    X_train, X_test, y_train, y_test = train_test_split(stats_desc, stats_label, test_size=0.2, random_state=42)

    # Créer un objet de l'algorithme d'arbre de décision
    clf = DecisionTreeClassifier()

    # Entraîner l'arbre de décision sur les données d'entraînement
    clf.fit(X_train, y_train)

    # Prédire les étiquettes des données de test
    predictions = clf.predict(X_test)

    # Calculer l'exactitude des prédictions
    accuracy = accuracy_score(y_test, predictions)
    print("Accuracy ", i, " : ", accuracy)
    
    sum_accuracy += accuracy
    
print (f'Accuracy moyenne : {sum_accuracy / 10}')
```

#### Utilisation du modèle implémenté en cours

```python
# Représentation de l'arbre de décision lorsque qu'il apprend l'intégralité du dataset

import graphviz as gv

seuil = 0.0

# Initialiation d'un arbre pour le dataset :
arbre = classif.ClassifierArbreNumerique(len(liste_features), seuil, liste_features)

# Construction de l'arbre de décision à partir du dataset entier 
arbre.train(stats_desc,stats_label)

# Construction de la représentation graphique (affichage)
graphe_arbre = gv.Digraph(format='png')
arbre.affiche(graphe_arbre)

# Affichage du graphe obtenu:
print("Cet arbre possède ",arbre.number_leaves()," feuilles.")
graphe_arbre
```

On observe que ces paramètres sont : % de victoires, nombre de points, rang de conférence, mais aussi (dans une moindre mesure) rang de division, nombre de passes décisives ou bien nombre de fautes personnelles.

Ces deux derniers paramètres sont intérresants car contrairement aux précédents, ceux ci ne conditionnent pas directement l'accès aux finales.

En effet, arriver en finales dépend directement du nombre de victoires/défaites ou du classement mais pas du nombre de d'assists et de fautes.

On peut juger qu'une équipe faisant de nombreuses passes décisives se repose davantage sur le collectif et possède une plus grande facilité à scorer, capacité impactant directement l'issue d'un match et influe donc sur le pourcentage de victoires.

### Régression linéaire

Essayons de prédire le nombre de victoires des Lakers sur une saison en fonction des performances des saisons précédentes.

```python
from sklearn.model_selection import train_test_split

# Définir les caractéristiques (features) et la cible (target)
features = ['FGM', 'FGA', 'FG_PCT', 'FG3M', 
            'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 
            'PF', 'STL', 'TOV', 'BLK', 'PTS', 'PTS_RANK']
target = 'WINS'

# Diviser les données en ensembles d'entraînement et de test
X = team_stats_df[features]
y = team_stats_df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Modèle de Régression Linéaire
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Modèle de Forêt Aléatoire
rf_model = RandomForestRegressor(n_estimators=100)
rf_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Prédictions sur l'ensemble de test
y_pred_lr = lr_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)

# Calculer les métriques de performance
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f"Régression linéaire - MSE: {mse_lr:.3f}, R2: {r2_lr:.3f}")
print(f"Forêt aléatoire - MSE: {mse_rf:.3f}, R2: {r2_rf:.3f}")
```

Observation : le modèle de forêt aléatoire possède un Mean Squared Error beaucoup plus faible que le modèle de régression linéaire. Cela montre que la forêt aléatoire s'ajuste mieux au données.

De plus, le R-squared du modèle de forêt est supérieure à celui du modèle de régression, montrant que le modèle explique mieux la variance des données.
